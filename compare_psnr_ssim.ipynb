```python
import json
with open('PSNR_SSIM_json_folder/baseline_SIREN_psnr_dict.json') as f:
    data_2 = json.load(f)
with open('PSNR_SSIM_json_folder/baseline_Fourier_psnr_dict.json') as f:
    data_1 = json.load(f)
with open('PSNR_SSIM_json_folder/baseline_SIREN_Fourier_psnr_dict.json') as f:
    data_3 = json.load(f)
```

```python
import matplotlib.pylab as plt

data_1 = {int(k):v for k,v in data_1.items()}
data_2 = {int(k):v for k,v in data_2.items()}
data_3 = {int(k):v for k,v in data_3.items()}

lists1 = sorted(data_1.items())
lists2 = sorted(data_2.items())
lists3 = sorted(data_3.items())

x1, y1 = zip(*lists1)
x2, y2 = zip(*lists2)
x3, y3 = zip(*lists3)

plt.plot(x1, y1, label='baseline')
plt.plot(x2, y2, label='SIREN')
plt.plot(x3, y3, label='SIREN+Fourier')
plt.legend()
plt.xlabel('image_num')
plt.ylabel('PSNR')
plt.show()
```
